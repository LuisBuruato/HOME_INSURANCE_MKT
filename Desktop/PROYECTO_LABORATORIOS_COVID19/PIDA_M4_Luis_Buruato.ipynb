```python
import pandas as pd

# Asegúrate de que la columna 'date' sea de tipo datetime
DatosFinalesFiltrado1['date'] = pd.to_datetime(DatosFinalesFiltrado1['date'], errors='coerce')

# Asegúrate de que la columna 'cumulative_vaccine_doses_administered' sea numérica
DatosFinalesFiltrado1['cumulative_vaccine_doses_administered'] = pd.to_numeric(DatosFinalesFiltrado1['cumulative_vaccine_doses_administered'], errors='coerce')

# Crear una columna de año y mes
DatosFinalesFiltrado1['year'] = DatosFinalesFiltrado1['date'].dt.year

# Filtrar los datos para Brasil y para el año 2021
datos_brasil_2021 = DatosFinalesFiltrado1[(DatosFinalesFiltrado1['country_name'] == 'Brazil') & (DatosFinalesFiltrado1['year'] == 2021)]

# Obtener la dosis total administrada en 2021
# Para obtener la cantidad total de dosis administradas durante 2021,
# calculamos la diferencia entre las dosis acumuladas al final y al principio del año.

# Tomamos el valor de las dosis al final de 2021 (última fila del año)
dosis_final_2021 = datos_brasil_2021[datos_brasil_2021['date'].dt.year == 2021]['cumulative_vaccine_doses_administered'].max()

# Tomamos el valor de las dosis al principio de 2021 (primer valor disponible)
dosis_inicio_2021 = datos_brasil_2021[datos_brasil_2021['date'].dt.year == 2021]['cumulative_vaccine_doses_administered'].min()

# La cantidad total de dosis administradas es la diferencia entre el valor máximo y mínimo de dosis acumuladas
total_dosis_2021 = dosis_final_2021 - dosis_inicio_2021

# Mostrar el resultado
print(f'La cantidad total de dosis administradas en Brasil en 2021 es: {total_dosis_2021:.0f} dosis')
```
